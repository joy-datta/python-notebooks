```
class Node:
    def __init__(self,value):
        self.prev=None
        self.next=None
        self.val=value

class DoubleLinkedList:
    def __init__(self):
        self.head=None
        self.tail=None
        self.size=0

    def add(self,value):
        node=Node(value)
        if self.tail is None:
            self.head=node
            self.tail=node
            self.size+=1
        else:
            self.tail.next=node
            node.prev=self.tail
            self.tail=node
            self.size+=1

    def remove(self,value):
        node=self.head
        while node is not None:
            if node.val==value:
                self.__remove(node)
                break
            node=node.next

    def __remove(self,node):
        if node.prev is None:
            self.head = node.next
        else:
            node.prev.next = node.next
        if node.next is None:
            self.tail = node.prev
        else:
            node.next.prev = node.prev
        self.size-=1

    def add_first(self,value):
        node=Node(value)
        self.head.prev=node
        node.next=self.head
        self.head=node
        self.size+=1

    def remove_last(self):
        self.tail=self.tail.prev
        self.tail.next=None
        self.size-=1
    def remove_first(self):
        self.head=self.head.next
        self.head.prev=None
        self.size-=1
    def __str__(self):
        nodes=[]
        node=self.head
        while node is not None:
            nodes.append(node.val)
            node=node.next
        return f"[{', '.join(str(val) for val in nodes)}]"

linked_list=DoubleLinkedList()
linked_list.add(1)
linked_list.add(3)
linked_list.add(4)
linked_list.add(7)
linked_list.add(8)
linked_list.add(9)
print(linked_list)

linked_list.remove(4)
print(linked_list)

linked_list.remove_first()
print(linked_list)

print(linked_list)
```

```
class Graph:
  def __init__(self,graph={}):
    self.graph=graph

  def return_dict_of_graph(self):
    return self.graph

  def addNode(node):   
    self.graph[node]=[]

  def addEdge(node,neighbor):
    self.graph[node].append(neighbor)
    self.graph[neighbor].append(node)

  def gen_edges(self):
    edges=[]   
    for node in self.graph:
      for neighbors in self.graph[node]:
        edges.append((node,neighbors))
    return edges    

  def nodeDel(self,node):
    del self.graph[node]        
    keys=list((self.graph).keys())
    for key in keys:
      for neighbors in self.graph[key]:
        if neighbors==node:
          self.graph[key].remove(neighbors)

  def edgeDel(self,node,neighbor):
    self.graph[node].remove(neighbor)
    self.graph[neighbor].remove(node)

  def isolated_nodes(self):  
    list_of_isolated_nodes=[]  
    for nodes in self.graph:
      number_of_neighbors=0 
      for neighbors in self.graph[nodes]:
        if neighbors:
          number_of_neighbors+=1
      if number_of_neighbors==0:
        list_of_isolated_nodes.append(nodes)
    return list_of_isolated_nodes

  def dergee_of_nodes(self):
    degrees_dict={}
    for key in list(self.graph.keys()):
      degrees_dict[key]=len(self.graph[key])
    return degrees_dict

  def connect_two_graphs(self,given_graph,node1,node2):
    given_graph=given_graph.return_dict_of_graph()
    self.graph[node1].append(node2)
    given_graph[node2].append(node1)
  
  def merge_two_graphs(self,given_graph):
    given_graph=given_graph.return_dict_of_graph()
    new_graph={}
    for graph in (self.graph,given_graph):
      for node in list(graph.keys()):      
        new_graph[node]=[]  
        for neighbors in graph[node]:
          new_graph[node].append(neighbors)
    return new_graph       


  
  def find_path(self,start,end,path=[]):
    path.append(start)
    if start==end:
      return path   
    else:
      for neighbors in self.graph[start]:
        if neighbors==end:
          path.append(neighbors)
          return path   
        else:
          new_path=find_path(self.graph,neighbors,end,path)
        if new_path:
          return new_path
        else:
          return None


g=Graph({'a':['b','c','k'],
         'b':['a','b','e','k'],
         'c':['a','h'],
         'n':[],
         'g':[],
         'e':['b','h'],
         'h':['c','e'],
         'k':['a','b']})
graph=Graph({'x':['x','z'],
             'y':['z'],
             'z':['x','y']})

g.connect_two_graphs(graph,'a','z')
g.connect_two_graphs(graph,'c','y')
print(g.gen_edges())
print(graph.gen_edges())
graph_x=Graph(g.merge_two_graphs(graph))
print(graph_x.return_dict_of_graph())
print(graph_x.find_path('x','h'))
```

```
#weighted undirected graph        
graph=[[8,10,None,12,14],
       [10,0,12,None,None],
       [None,12,0,None,14],
       [12,None,None,0,8],
       [14,None,14,8,0]]
#find which edges are connected with node 0?        
def connected_nodes_with_node(node):
  edges=[]
  for weights in graph[node]: 
    if weights is not None or weights!=0:       
      edges.append((graph[node].index(weights),weights))
  return edges 
def gen_edges(graph):
  edges=[]
  for nodes in graph:
    index_of_graph=graph.index(nodes)
    for weights in nodes:
      edges.append((index_of_graph,(graph[index_of_graph]).index(weights),weights))
  return edges
print(connected_nodes_with_node(0))
print(gen_edges(graph))
```

```
#UNWEIGHTED & UNDIRECTED GRAPH
graph={1:[3,4],
       2:[4,5],
       3:[1,4,7],
       4:[1,2,3,5,7,8],
       5:[2,6,8],
       6:[5,8],
       7:[3,4,8,9,11],
       8:[4,5,6,7,9,10,11],
       9:[7,8,11],
       10:[7,8,11],
       11:[7,8,9,10],
       12:[]}
def add_node(graph,node):
  graph[node]=[]
def add_edge(graph,node,neighbor):
  graph[node].append(neighbor)
  graph[neighbor].append(node) #for directed graph this line is not required 
def gen_edges(graph):
  edge_list=[]
  for node in graph:
    for neighbors in graph[node]:
      edge_list.append((node,neighbors))
  return edge_list
def find_path(graph,start,end,path=[]):
  path.append(start)
  if start==end:
    return path
  for neighbors in graph[start]:
    if neighbors==end:
      path.append(neighbors)
      return path
  for neighbors in graph[start]:
    if neighbors not in path:
      new_path=find_path(graph,neighbors,end,path)
      if new_path:
        return new_path
      else:
        return None 
print(find_path(graph,1,10))
print(gen_edges(graph))
```
